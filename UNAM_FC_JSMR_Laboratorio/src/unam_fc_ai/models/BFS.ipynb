# BFS

BFS es fundamental no solo para grafos, sino como base para agentes que buscan metas en espacios de estados (como resolver un cubo Rubik o encontrar la ruta más corta en un mapa sin pesos). BFS garantiza encontrar el camino más corto en grafos donde todas las aristas valen lo mismo. En inteligencia artificial, esto se conoce como un algoritmo de búsqueda completo (siempre encuentra la solución si existe) y óptimo.


```python
from collections import deque
```

```python
def bfs(grafo, inicio):
    # Conjunto para rastrear nodos visitados
    visitados = set()
    # Cola para manejar el orden de exploración (FIFO)
    cola = deque([inicio])

    # Marcamos el inicio como visitado
    visitados.add(inicio)

    print(f"Orden de exploración empezando desde {inicio}:")

    while cola:
        # Extraemos el nodo que llegó primero
        nodo_actual = cola.popleft()
        print(nodo_actual, end=" ")

        # Revisamos los vecinos del nodo actual
        for vecino in grafo[nodo_actual]:
            if vecino not in visitados:
                visitados.add(vecino)
                cola.append(vecino)
```

```python
# Representación del grafo mediante un diccionario (Lista de adyacencia)
grafo_ejemplo = {
    0: [1, 2],
    1: [2],
    2: [0, 3],
    3: [3]
}
```

```python


# Ejecución
bfs(grafo_ejemplo, 2)
```
