# PARTE 3: Torres de Hanoi

Mover 3 discos desde la torre A hasta la torre C usando B como auxiliar.


Reglas:
1) Solo mover un disco a la vez.
2) No colocar un disco grande sobre uno pequeño.



Escriba un programa que simule los pasos necesarios para resolver el rompecabezas de la "Torre de Hanoii" para algunos discos n.
- https://www.mathsisfun.com/games/towerofhanoi.html

- Algoritmo recursivo
- Si hay 1 o más discos para mover:
     1. Mueva los discos n-1 superiores de la aguja 1 (fuente) a la aguja 2 (ayudante), usando la aguja 3 (destino) como aguja intermedia
     2. Mueva el disco número n de la aguja 1 a la aguja 3
     3. Mueva los discos n - 1 superiores de la aguja 2 a la aguja 3, usando la aguja 1 como aguja intermedia.

```python
def moveDisks(n, src, helper, dst):
    if n > 0:
        moveDisks(n-1, src, dst, helper)
        print('Move disk #{} from {} to {}'.format(n, src, dst))
        moveDisks(n-1, helper, src, dst)
```
